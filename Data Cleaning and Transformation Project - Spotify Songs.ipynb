```python
#Specifying the column names that I want to use.
new_cols_dict = {
    'trackName':'track',
    'artistName':'artist',
    'msPlayed':'minutes',
    'genre':'genre',
    'danceability':'danceability',
    'energy':'energy',
    'key':'key',
    'loudness':'loudness',
    'mode':'mode',
    'speechiness':'speechiness',
    'acousticness':'acousticness',
    'instrumentalness':'instrumentalness',
    'liveness':'liveness',
    'valance':'valance',
    'tempo':'tempo',
    'type':'type',
    'id':'id',
    'uri':'uri',
    'track_href':'track_href',
    'analysis_url':'analysis_url',
    'duration_ms':'duration',
    'time_signature':'time_signature'
}
    
#Assigning the new names to columns using rename.
spotify.rename(new_cols_dict, axis =1, inplace = True)
```

```python
spotify.columns
```

```python
#Converting the duration column from milliseconds to seconds. 
spotify.duration = (spotify.duration/1000)
```

```python
#Converting the minutes column to hours. 
spotify.minutes = (spotify.minutes/60)
```

```python
#Renaming the minutes column to 'hours'.
spotify.rename(columns = {'minutes':'hours'}, inplace= True)
```

```python
#Rounding each float type in the DataFrame to 2 decimal places. This gives us a cleaner look at the data. 
spotify = spotify.round(2)
```

```python
spotify
```

```python
#If we wanted to perform analysis on the 'key' column, it would be useful to rename the values to their proper key names. 
#Creating a dictionary to map the values in "key" column to actual musical key names.
key_dict = {
    '0.0':'C',
    '1.0': 'Db/C#',
    '2.0': 'D',
    '3.0': 'Eb',
    '4.0': 'E',
    '5.0' : 'F',
    '6.0': 'Gb/F#',
    '7.0' : 'G',
    '8.0': 'Ab',
    '9.0': 'A',
    '10.0' : 'Bb',
    '11.0' : 'B/Cb'
}

#Before we map, we must change the data type of the 'key' column to string. 
spotify.key = spotify.key.astype('str')
spotify.key

#Use the map function to complete the mapping. 
spotify.key = spotify.key.map(key_dict)
spotify.key
```

```python
spotify #View the DataFrame with the updated 'key' column. 
```

```python
#Remove unneccessary columns from the dataset. 
spotify = spotify.drop(columns = ["type", "id","uri","track_href","analysis_url","time_signature"])
```

```python
#Writing final dataframe to an excel file. 
spotify.to_excel('spotify_final.xlsx', sheet_name = 'Data')
```
